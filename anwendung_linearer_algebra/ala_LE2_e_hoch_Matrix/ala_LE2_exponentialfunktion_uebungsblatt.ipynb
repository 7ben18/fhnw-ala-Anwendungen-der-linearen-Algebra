```python
# e^x approximieren mit Taylor Polynom Funktion. 
x = np.linspace(-2,2,1000)
plt.figure(figsize=(10,10))
plt.plot(x, np.exp(x), label = "$e^x$")
plt.plot(x, np.ones_like(x), label = "$1$")
plt.plot(x, 1 + x, label = "$1 + x$")
plt.plot(x, 1 + x + 1/2 * x**2, label = "$1 + x + 1/2 * x^2$")
plt.plot(x, 1 + x + 1/2 * x**2 + 1/6 * x ** 3, label = "$1 + x + 1/2 * x^2 + 1/6 * x^3")
plt.plot(x, 1 + x + 1/2 * x**2 + 1/6 * x ** 3 + 1/24 * x ** 4, label = "$1 + x + 1/2 * x^2 + 1/6 * x^3 + 1/24 * x^4")
plt.legend()
plt.show()
```

```python
# 1. Exponentieren wir die Nullmatrix, so erhalten wir die Einheitsmatrix: e^0 = I.
Zero_Matrix = np.array([[0,0],[0,0]]) 

Einheitsmatrix = expm(Zero_Matrix)
print(Einheitsmatrix)
```

```python
# 2 Die Transponierte vertauscht:
A = np.array([[1,2],[3,4]])
expm_A = expm(A)
expm_A_T = expm_A.T
print(expm_A_T)

A_T = A.T
expm_A_T2 = expm(A_T)
print(expm_A_T2)

# Ob ich die zuerst e^Matrix oder e^Transponierte Matrixe mache spielt keine Rolle, beides ergibt das gleiche.
```

```python
# 3 
A = np.array([[1,2],[3,4]])
B = np.array([[3,4],[2,1]])

expm_AB = expm(A + B)
expm_AB
```

```python
M = np.array([[1,2],[2,1]])
print(M)

eig, T = np.linalg.eig(M)
print(eig)
print(T)
```

```python
T_inverse = np.linalg.inv(T)
A = np.diag(eig)
```

```python
expm(T @ A @ T_inverse)
```

```python
T @ expm(A) @ T_inverse
```

```python
import numpy as np
```

# Wie rechne ich eine Matrix e^A aus?

```python
# Definieren uns die Matrix A
A = np.array([[0,5],[-1,6]])
print(A)
```

# 1) Diagonalisierung der Matrix A = SDS^-1

```python

eigenwert, eigenvektoren = np.linalg.eig(A)
print(eigenwert)
print(eigenvektoren)

# Definieren der Matrix D mittels Eigenwerte
D = np.diag(eigenwert)
print(D)

S = eigenvektoren
print(eigenvektoren)

S_inv = np.linalg.inv(S)
print(S_inv)

# Testen ob SDS^-1 = A ist
print(np.round(S @ D @ S_inv))
```

# 2) Formel der Berechnung aufstellen:

# 2) S und S^-1 haben wir herausgefunden, nun suchen wir e^D, D kennen wir!

```python
# e^D definieren
eD = np.exp(eigenwert)

eD = np.diag(eD)
print(eD)
```

# Formel fuer e^A = S * eD * S^-1

```python
eA = S @ eD @ S_inv
print(eA)
```

# Ueberpruefen ob die Formel stimmt

```python
from scipy.linalg import expm

eA_check = expm(A)
print(eA_check)
```

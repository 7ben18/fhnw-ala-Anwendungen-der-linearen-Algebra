```python
import numpy as np
```

```python
# Singulaerwertzerlegung => A = U Z V 
# U und V muessen Orthogonal sein => U @ U.T = I 
A = np.array([[3,0],[4,5]])
display(A)

U = (1/20**0.5) * np.array([[1,-3], [3, 1]])
display(U)

Z = (1/5**0.5) * np.array([[3,0],[0,5]])
display(Z)

V = np.array([[1,-1], [1,1]])
display(V) 

# Testen ob V.T @ V = I
display(V.T @ V)
# -> Keine Einheitsmatrix 
# => Keine Singulaerwertzerlegung!


# Rang bestimmen 
A_rang = np.linalg.matrix_rank(A)
U_rang = np.linalg.matrix_rank(U)
Z_rang = np.linalg.matrix_rank(Z)
V_rang = np.linalg.matrix_rank(V)
print(A_rang, U_rang, Z_rang, V_rang)
```

```python
A = np.array([[4,0],[3,-5]])
```

```python
# Funktion definieren
def get_Z(A: np.array):
    """
    Input: Matrix A
    output: Matrix Z
    """
    ATA = A.T @ A
    eigw, eigv = np.linalg.eig(ATA)
    Z = np.diag(np.sqrt(sorted(eigw, reverse = True)))
    return Z
```

```python
display(get_Z(A))
```

```python
A = np.array([[4,0],[3,-5]])
AT = A.T
```

```python
display(get_Z(A), get_Z(AT))
```

```python
A_inv = np.linalg.inv(A)
A_inv
```

```python
display(get_Z(A_inv))
```

```python
def get_singulaer_values(A: np.array):
    """
    Input: Matrix A
    output: singularvalues
    """
    ATA = A.T @ A
    eigw, eigv = np.linalg.eig(ATA)
    singualervalue = (np.sqrt(sorted(eigw, reverse = True)))
    return singualervalue

```

```python
A_singular_value = get_singulaer_values(A)
display(A_singular_value)

print( 1 / A_singular_value )
```

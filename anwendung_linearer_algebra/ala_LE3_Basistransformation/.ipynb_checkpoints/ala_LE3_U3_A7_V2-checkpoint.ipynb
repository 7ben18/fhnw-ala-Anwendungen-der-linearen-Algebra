```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [-2, -4, -6, 6, -8, 6, 3, 6, 8, -9, -1, 1, 4, 2, 2, 2, -1, -1, 2, 2]
y = [-3, -7, -13, 16, -14, 14, 6, 11, 18, -19, 5, 0, 3, 3, 6, 8, -3, -1, 0, 9]

X = np.array([x,y])
X
```

```python
# Eigenwerte und Eigenvektoren von XXT
eigenwerte, eigenvektoren = np.linalg.eig(X @ X.T)
print(eigenwerte)
print(eigenvektoren)
```

```python
eigenvektoren = eigenvektoren[:, ::-1]
```

```python
T = np.linalg.inv(eigenvektoren)
T
```

```python
XT = T @ X
print(XT[0:1, : ])
print("")
print(XT[1:2, : ])
print("")
print(XT)
```

```python
# Eigenvektor als X Achse transofrmiert.
# Hauptkoompenten analyse
# Eigenvektore berechenn X XT, nach eigenwerte sortieren, -> 1. Hauptkompontenten etc... 
# Varianz der daten in 2 Dimensionen darstellen
# Viel vereaenderungen = Varianz
```

```python
plt.scatter(X[0, :], X[1, :], c = "b")
plt.scatter(XT[0, :], XT[1, :], c = "r")
plt.show()
```

```python
# die blauen Punkte werden auf X-Achse transformiert. 
# Transoformation in die Eigenbasis.
# Varianz = Veranederung ist eine grosse Zahl.
# Hauptkompontent Analyse - PCA
```

```python
# Durch die Standardisierung kann mir die Einheit egal sein! 
```

```python
x = [-2, -4, -6, 6, -8, 6, 3, 6, 8, -9, -1, 1, 4, 2, 2, 2, -1, -1, 2, 2]
y = [-3, -7, -13, 16, -14, 14, 6, 11, 18, -19, 5, 0, 3, 3, 6, 8, -3, -1, 0, 9]

x = (np.array(x) - np.mean(x)) / np.std(x)
y = (np.array(y) - np.mean(y)) / np.std(y)

X = np.array([x,y])
X
```

```python
x_mean_test, x_std_test = np.mean(X[0,:]), np.std(X[0,:])
y_mean_test, y_std_test = np.mean(X[1,:]), np.std(X[1,:])
print(x_mean_test, x_std_test, y_mean_test, y_std_test)
```

```python
eigenwerte, eigenvektoren = np.linalg.eig(X @ X.T)
print(eigenwerte)
print(eigenvektoren)
```

```python
eigenwerte = eigenwerte[::-1]
eigenwerte
```

```python
eigenvektoren = eigenvektoren[:,::-1]
eigenvektoren
```

```python
T = np.linalg.inv(eigenvektoren)
T
```

```python
XT = T @ X
XT
```

```python
plt.scatter(X[0,:], X[1,:], c = "b")
plt.scatter(XT[0,:], XT[1,:], c = "r")
plt.show()
```

```python
# Module importieren
import numpy as np
import pandas as pd
```

```python
# Daten in einlesen
data = pd.read_csv("sales_subset.csv")
data
```

```python
# Daten selektieren
data_select = data[["weekly_sales", "temperature_c", "fuel_price_usd_per_l", "unemployment"]]
data_select
```

```python
# b Vektor definieren
b = np.array(list(data_select["weekly_sales"])).reshape(-1,1)


# A Vektoren definieren
A_temp = np.array(list(data_select["temperature_c"])).reshape(-1,1)
A_fuel = np.array(list(data_select["fuel_price_usd_per_l"])).reshape(-1,1)
A_unemp = np.array(list(data_select["unemployment"])).reshape(-1,1)
A_ones = np.ones(data_select.shape[0]).reshape(-1,1)

# A Matrix definieren
A = np.hstack([A_ones, A_temp, A_fuel, A_unemp])
```

```python
# x Parameter herausfinden
x = np.linalg.inv(A.T @ A) @ A.T @ b
print(x)

data_select["weekly_sales_predict"] = x[0] + x[1] * data_select["temperature_c"] + x[2] * data_select["fuel_price_usd_per_l"] + x[3] * data_select["unemployment"]
data_select
```

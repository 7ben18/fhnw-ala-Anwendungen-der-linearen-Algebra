```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[89., 75., 22., 102.],[75., 116., 27., 120.],[22., 27., 33., 62.], [102., 120., 62., 200.]])
A
```

```python
#a 
eigw, eigv = np.linalg.eig(A)
display(eigw, eigv)
```

```python
#b 
np.round(eigv @ eigv.T)
# Weil die Matrix A symetrisch ist
```

```python
A = np.array([[1,2],[2,1]])
B = np.array([[1,-1],[-1,4]])
C = np.array([[1,0,0],[0,1,1],[0,1,1]])
D = np.array([[2,-1,0],[-1,2,-1],[0,-1,2]])
```

```python
def positiv_definit(A: np.array):
    return (np.linalg.eigvals(A) > 0).all()
```

```python
Matrices = [A, B, C, D]
for M in Matrices:
    print(positiv_definit(M))
```

```python
x = np.array([-2, -4, -6, 6, -8, 6, 3, 6, 8, -9, -1, 1, 4, 2, 2, 2, -1, -1, 2, 2])
y = np.array([-3, -7, -13, 16, -14, 14, 6, 11, 18, -19, 5, 0, 3, 3, 6, 8, -3, -1, 0, 9])
X = np.hstack([[x,y]])
X
```

```python
XXT = X @ X.T 
eigw, eigv = np.linalg.eig(XXT)
eigw = sorted(eigw, reverse = True)
eigw = np.diag(eigw)
eigv = eigv[ : , ::-1]
display(eigw, eigv)
```

```python
T = eigv
T
```

```python
T_inv = np.linalg.inv(T)
```

```python
X_transform = T_inv @ X
X_transform
```

```python
plt.scatter(X[ 0 , :], X[ 1 , :], c = "red", label = "Vor Transformation")
plt.scatter(X_transform[ 0 , :], X_transform[ 1 , :], c = "b", label = "Nach Transformation")
plt.grid()
plt.legend()
plt.show()
```

```python
x_n = np.array([-2, -4, -6, 6, -8, 6, 3, 6, 8, -9, -1, 1, 4, 2, 2, 2, -1, -1, 2, 2])
y_n = np.array([-3, -7, -13, 16, -14, 14, 6, 11, 18, -19, 5, 0, 3, 3, 6, 8, -3, -1, 0, 9])

x_n = (x_n - x_n.mean()) / x_n.std()
y_n = (y_n - y_n.mean()) / y_n.std()

X_S = np.hstack([[x_n, y_n]])
X_S
```

```python
XSXST = X_S @ X_S.T 
XSXST
```

```python
eigw, eigv = np.linalg.eig(XSXST)
display(eigw, eigv)
```

```python
eigw = np.diag(sorted(eigw, reverse = True))
eigw
```

```python
eigv = eigv[: , ::-1]
eigv
```

```python
T = eigv
T
```

```python
T_inv = np.linalg.inv(T)
```

```python
X_S_Transform = T_inv @ X_S
X_S_Transform
```

```python
plt.figure(figsize = (20,10))
plt.scatter(X[ 0 , :], X[ 1 , :], c = "red", label = "Vor Transformation")
plt.scatter(X_transform[ 0 , :], X_transform[ 1 , :], c = "b", label = "Nach Transformation")
plt.scatter(X_S[ 0 , :], X_S[ 1 , :], color = "orange", label = "Vor Transformation Standardisiert")
plt.scatter(X_S_Transform[ 0 , :], X_S_Transform[ 1 , : ], color = "cyan", label = "Nach Transformation Standardisiert")
plt.grid()
plt.legend()
plt.show()
```
